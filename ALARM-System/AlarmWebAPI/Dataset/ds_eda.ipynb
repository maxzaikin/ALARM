### Fill NA

```python
def remove_anomaly(x):
    x = x.str.replace(r'\xa0', '', regex=True).str.strip()
    x = x.str.replace(',', '.', regex=False).str.replace('-', '0', regex=False)     
    
    return x
    

    
df['stock_size']= remove_anomaly(df['stock_size'])
df['stock_size']= df['stock_size'].fillna('0.0')

df['address']= df['address'].fillna('')
df['ceo_fn']= df['ceo_fn'].fillna('')
df['ceo_bdt']= df['ceo_bdt'].fillna('1/1/1900')
df['beneficiary_fn']= df['beneficiary_fn'].fillna('')
df['prim_okved']= df['prim_okved'].fillna('0.00')
df['okved_count']= df['okved_count'].fillna(-1)
df['employee_count']= df['employee_count'].fillna(-1)
df['website']= df['website'].fillna('')

df['phone_number']= df['phone_number'].fillna('')
df['phone_number'] = df['phone_number'].astype(str).str.replace(r'\.0$', '', regex=True)

df['phone_operator']= df['phone_operator'].fillna('')
df['taxation_system']= df['taxation_system'].fillna('')
df['client_activity_checked']= df['client_activity_checked'].fillna('')
df['client_activity_verb']= df['client_activity_verb'].fillna('')
df['employee_count_verb']= df['employee_count_verb'].fillna(-1)

df['planned_income_verb']= remove_anomaly(df['planned_income_verb'])
df['planned_income_verb']= df['planned_income_verb'].fillna('0.0')

df['planned_withdrawal_verb']= remove_anomaly(df['planned_withdrawal_verb'])
df['planned_withdrawal_verb']= df['planned_withdrawal_verb'].fillna('0.0')

df['income']= remove_anomaly(df['income'])
df['income']= df['income'].fillna('0.0')

df['zsk']= df['zsk'].fillna(-1)
df['is_negative_info']= df['is_negative_info'].fillna(0)
df['is_negative_ceo_info']= df['is_negative_ceo_info'].fillna(0)
df['is_fraudelent']= df['is_fraudelent'].fillna(0)
df['reg_services']= df['reg_services'].fillna(-1)
df['sim_card_lifetime_since_change']= df['sim_card_lifetime_since_change'].fillna('1/1/1900')
df['sim_lifetime_in_current_device']= df['sim_lifetime_in_current_device'].fillna('1/1/1900')
df['sim_card_lifetime_since_contract']= df['sim_card_lifetime_since_contract'].fillna('1/1/1900')

df['tax_burden']= df['income']= remove_anomaly(df['tax_burden'])
df['tax_burden']= df['tax_burden'].fillna(0.0)
```

### Fix types

```python
df['inn']= df['inn'].astype('object')

df['reg_dt']= pd.to_datetime(df['reg_dt'], errors='coerce')
#df['reg_dt'] = df['reg_dt'].fillna(pd.Timestamp('9999-09-09'))

df['stock_size']= df['stock_size'].astype('float64')

df['ceo_bdt']= pd.to_datetime(df['ceo_bdt'], errors='coerce')
#df['ceo_bdt'] = df['ceo_bdt'].fillna(pd.Timestamp('9999-09-09'))

df['okved_count']= df['okved_count'].astype('int')
df['employee_count']= df['employee_count'].astype('int')
df['employee_count_verb']= df['employee_count_verb'].astype('int')
df['planned_income_verb']= df['planned_income_verb'].astype('float64')
df['phone_number']= df['phone_number'].astype('object')
df['planned_withdrawal_verb']= df['planned_withdrawal_verb'].astype('float64')
df['income']= df['income'].astype('float64')
df['zsk']= df['zsk'].astype('int')

df['is_negative_info']= df['is_negative_info'].apply(lambda x: 1 if x=='Имеется' else 0)
df['is_negative_info']= df['is_negative_info'].astype('int')

df['is_negative_ceo_info']= df['is_negative_ceo_info'].apply(lambda x: 1 if x=='Имеется' else 0)
df['is_negative_ceo_info']= df['is_negative_ceo_info'].astype('int')

df['is_fraudelent']= df['is_fraudelent'].apply(lambda x: 1 if x=='Имеется' else 0)
df['is_fraudelent']= df['is_fraudelent'].astype('int')

df['reg_services']= df['reg_services'].astype('float64')

df['sim_card_lifetime_since_change']= pd.to_datetime(df['sim_card_lifetime_since_change'], errors='coerce')
df['sim_card_lifetime_since_change'] = df['sim_card_lifetime_since_change'].fillna(pd.Timestamp('1900-1-1'))

df['sim_lifetime_in_current_device']= pd.to_datetime(df['sim_lifetime_in_current_device'], errors='coerce')
df['sim_lifetime_in_current_device'] = df['sim_lifetime_in_current_device'].fillna(pd.Timestamp('1900-1-1'))

df['sim_card_lifetime_since_contract']= pd.to_datetime(df['sim_card_lifetime_since_contract'], errors='coerce')
df['sim_card_lifetime_since_contract'] = df['sim_card_lifetime_since_contract'].fillna(pd.Timestamp('1900-1-1'))

df['tax_burden']= df['tax_burden'].astype('float64')
```

```python
def analyze_inn(inn):
    if not isinstance(inn, str):
        return 0 # Incorrect  
    
    inn = inn.strip()
    
    if len(inn) == 10:
        return 1 # Company
    elif len(inn) == 12:
        return 2 # Physical 
    else:
        return 0 # Incorrect
        
df['inn'] = df['inn'].astype(str)        
df['inn_type'] = df['inn'].apply(analyze_inn)
```

```python
def categorize_registration_date(rd):
   
    if pd.isna(rd):
        return 'error'

    today = datetime.today()
    delta = today - rd

    if delta.days < 365:
        return 'new'
    elif 365 <= delta.days < 1095:
        return 'mid'
    else:
        return 'old'

# Пример применения к столбцу DataFrame
df['company_age'] = df['reg_dt'].apply(categorize_registration_date)
```

```python
def remove_last_four_digits(phone_number):    
    if pd.isna(phone_number):
        return phone_number      
    
    phone_number = str(phone_number).strip()
    return phone_number[:-4] if len(phone_number) > 4 else ''

df['phone_mask'] = df['phone_number'].apply(remove_last_four_digits)
```

```python
columns_to_merge = [
    'is_negative_info', 
    'is_negative_ceo_info', 
    'is_fraudelent', 
    'reg_services'
]

df['reg_services'] = df['reg_services'].apply(lambda x: 1 if x > 0.5 else 0)
df['is_fraud'] = df[columns_to_merge].max(axis=1)
```
